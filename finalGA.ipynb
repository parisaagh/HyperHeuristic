**Problem 1: Optimum Design of an I-beam**

```r
# genetic algorithms
install.packages('GA')

# statistical graphics
install.packages('ggplot2')

# used to create plots and charts, including line charts
# scatter plots, bar charts , histograms, etc.
install.packages('plotly')
```

```r
# Loading the needed libraries

library(GA)
library(plotly)
```

```r
f1 <- function(x) {
  return(2 * x[2] * x[4] + x[3] * (x[1] - 2 * x[4]))
}


f2 <- function(x) {
  return(60000 / (x[3] * (x[1] - 2 * x[4])^3 + 2 * x[2] * x[4] * (4 * x[4]^2 + 3 * x[1] * (x[1] - 2 * x[4]))))
}
```

```r
# Define the fitness function considering the weighted approach
fitness <- function(x) {
  a <- 0.7  # Weight for f1
  b <- 0.3  # Weight for f2
  return(a * f1(x) + b * f2(x))
}

# GA parameters
numGenerations <- 100
populationSize <- 50
Pc <- 0.75
Pm <- 0.001

# Define the lower and upper bounds for variables
lower_bound <- c(10, 10, 0.9, 0.9)
upper_bound <- c(80, 50, 5, 5)

# Run the GA
GA_result <- ga(type = "real-valued",
                fitness = fitness,
                lower = lower_bound,
                upper = upper_bound,
                popSize = populationSize,
                pcrossover = Pc,
                pmutation = Pm,
                maxiter = numGenerations,
                run = 100,
                keepBest = TRUE,
                seed = 123,
                monitor = TRUE)

summary(GA_result)

```

```r
# Visualize convergence for f2 and plot the best solution for each generation
plot(GA_result)
for (i in 1:length(GA_result@solution)) {
  points(rep(i, length(GA_result@solution[[i]])), GA_result@solution[[i]], col = 'red', pch = 20)
}
```

```r
# Extract the best solution
best_solution <- GA_result@solution[1, ]

# Evaluate the best solution
f1_best <- f1(best_solution)
f2_best <- f2(best_solution)

# Print the best solution and its objective function values
cat("Best Solution:", best_solution, "\n")
cat("Best f1(X) =", f1_best, "\n")
cat("Best f2(X) =", f2_best, "\n")
```

```r
# Run the GA for f1
GA_result_f1 <- ga(type = "real-valued",
                fitness = f1,
                lower = lower_bound,
                upper = upper_bound,
                popSize = populationSize,
                pcrossover = Pc,
                pmutation = Pm,
                maxiter = numGenerations,
                run = 100,
                keepBest = TRUE,
                seed = 123,
                monitor = TRUE)

# Summary of GA for f1
summary(GA_result_f1)

# Visualize convergence for f1
plot(GA_result_f1)
```

```r
# Run the GA for f2
GA_result_f2 <- ga(type = "real-valued",
                fitness = f2,
                lower = lower_bound,
                upper = upper_bound,
                popSize = populationSize,
                pcrossover = Pc,
                pmutation = Pm,
                maxiter = numGenerations,
                run = 100,
                keepBest = TRUE,
                seed = 123,
                monitor = TRUE)

# Summary of GA for f2
summary(GA_result_f2)

# Visualize convergence for f2
plot(GA_result_f2)
```

```r
library(ggplot2)

# Extract fitness values from the monitor slot for f1
data_f1 <- data.frame(Generation = 1:numGenerations,
                      Best_Fitness = GA_result_f1@fitnessValue,
                      Avg_Fitness = rowMeans(GA_result_f1@population))

# Extract fitness values from the monitor slot for f2
data_f2 <- data.frame(Generation = 1:numGenerations,
                      Best_Fitness = GA_result_f2@fitnessValue,
                      Avg_Fitness = rowMeans(GA_result_f2@population))

# Plotting for f1
ggplot(data_f1, aes(x = Generation)) +
  geom_line(aes(y = Best_Fitness, color = "Best Fitness (f1)"), linetype = "solid") +
  geom_line(aes(y = Avg_Fitness, color = "Average Fitness (f1)"), linetype = "dashed") +
  labs(title = "Convergence of Fitness Function (f1)", x = "Generation", y = "Fitness Value") +
  scale_color_manual(values = c("Best Fitness (f1)" = "red", "Average Fitness (f1)" = "blue")) +
  theme_minimal()

# Plotting for f2
ggplot(data_f2, aes(x = Generation)) +
  geom_line(aes(y = Best_Fitness, color = "Best Fitness (f2)"), linetype = "solid") +
  geom_line(aes(y = Avg_Fitness, color = "Average Fitness (f2)"), linetype = "dashed") +
  labs(title = "Convergence of Fitness Function (f2)", x = "Generation", y = "Fitness Value") +
  scale_color_manual(values = c("Best Fitness (f2)" = "red", "Average Fitness (f2)" = "blue")) +
  theme_minimal()
```
